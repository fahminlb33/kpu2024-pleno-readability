```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_parquet("../data/clean/summaries.parquet")
df.head()
```

```python
df["suara_masuk_persen"].hist(bins=100)
```

```python
# df_subset = df[df["kode"].str.len() == 6][["suara_paslon_1", "suara_paslon_2", "suara_paslon_3"]].copy().dropna()
df_subset = df[df["kode"].str.len() == 6].copy().dropna()
df_subset
```

```python
def benford_proportion(digit):
    return np.log10((digit+1)/digit)

def calculate_proportions(series: pd.Series):
    dfp = series.copy().astype(int).astype(str).str[0].value_counts().sort_index().reset_index()
    dfp.columns = ["digit", "count"]

    dfp = dfp[dfp["digit"] != "0"]
    dfp["actual"] = dfp["count"] / dfp["count"].sum()
    dfp["benford"] = [benford_proportion(digit) for digit in range(1, 10)]
    
    return dfp
```

```python
df_proportions = calculate_proportions(df_subset["suara_paslon_3"])
df_proportions
```

```python
fig, ax = plt.subplots()

ax.bar(df_proportions["digit"], df_proportions["actual"], label="Actual proportion")
ax.plot(df_proportions["digit"], df_proportions["benford"], color="r", label="Proportion according to Benford's law", marker="o")
ax.set_xlabel("first_digit")
ax.set_ylabel("Likelihood of being first digit")

plt.legend()
plt.show()
```

## Semua Paslon

```python
fig, ax = plt.subplots()

colors = ["red", "orange", "limegreen"]
for i in range(1, 4):
    df_proportions = calculate_proportions(df_subset[f"suara_paslon_{i}"])

    if i == 1:
        ax.bar(df_proportions["digit"], df_proportions["benford"], label="Proportion according to Benford's law")
        ax.set_xlabel("First digit")
        ax.set_ylabel("Likelihood of being first digit")
    
    ax.plot(df_proportions["digit"], df_proportions["actual"], marker="o", color=colors[i-1], label=f"Actual proportion paslon {i}")

ax.set_title("Benford' Law of Indonesian 2024 General Presiential Election [Kecamatan level]")

plt.legend()
plt.show()
```
